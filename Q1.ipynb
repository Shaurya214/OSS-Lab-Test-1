```python
import numpy as np
import matplotlib.pyplot as plt
lions=np.array([15,16,17,20,19,21,23,24,25,27])
elephants=np.array([50,52,54,53,55,56,57,59,60,62])
zebras=np.array([100,98,95,97,96,94,95,93,92,90])
species_data={'Lions':lions,'Elephants':elephants,'Zebras':zebras}
def calculate_total_population(species_data):
    return {species:np.sum(data) for species,data in species_data.items()}
def calculate_avg_yearly_growth(species_data):
    avg_growth={}
    for species,data in species_data.items():
        yearly_growth=np.diff(data)
        avg_growth[species]=np.mean(yearly_growth)
    return avg_growth
def calculate_yearly_growth_rate(species_data):
    growth_rate={}
    for species,data in species_data.items():
        yearly_growth_rate=np.diff(data)/data[:-1]*100
        growth_rate[species]=yearly_growth_rate
    return growth_rate
def visualize_population_trends(species_data):
    years=np.arange(1,11)
    plt.figure(figsize=(10,6))
    for species,data in species_data.items():
        plt.plot(years,data,label=species)
        plt.xlabel('Year')
        plt.ylabel('Population in thousands')
        plt.title('Population trends over 10 years')
        plt.legend()
        plt.grid(True)
        plt.show()
def find_highest_avg_growth(avg_yearly_growth):
    return max(avg_yearly_growth,key=avg_yearly_growth.get)

def create_population_bar_chart(species_data):
    final_year_pop={species:data[-1] for species,data in species_data.items()}
    plt.figure(figsize=(8,6))
    plt.bar(final_year_pop.keys(),final_year_pop.values(),color=['orange','green','blue'])
    plt.xlabel('Species')
    plt.ylabel('Population in Year 10 in thousands')
    plt.title('Total Population at year 10')
    plt.show()
    
total_population=calculate_total_population(species_data)
avg_yearly_growth=calculate_avg_yearly_growth(species_data)
yearly_growth_rate=calculate_yearly_growth_rate(species_data)

print("Total population over 10 years for each species:",total_population)
print("Average yearly population growth for each species:",avg_yearly_growth)

visualize_population_trends(species_data)

highest_avg_growth_species=find_highest_avg_growth(avg_yearly_growth)
print(f"The species with highest avg yearly growth rate is:{highest_avg_growth_species}")
create_population_bar_chart(species_data)
```
